### problem 5-3
Bipartite Graph Detection: We apply DFS to recursively color the vertices as black and white. A vertex should have a different color from its directly connected vertices have. If coloring fails, it's impossible to divide them into two parties. Time complexity is $O(|V| + |E|) = O(n)$.
